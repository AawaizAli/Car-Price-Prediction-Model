```python
# Remove the 'PKR' and 'lacs' from the Price column and handle rows with 'Call for price'
data['Price'] = data['Price'].str.replace('PKR ', '').str.replace(' lacs', '')
data = data[data['Price'] != 'Call for price']

# Convert cleaned price data to float, assuming 'lacs' means 'hundred thousands'
data['Price'] = data['Price'].astype(float) * 100000

# Show the cleaned data
cleaned_data_preview = data.head()
cleaned_data_preview
```

```python
# Remove 'cc' from the Displacement column and convert to integer type
data['Displacement'] = data['Displacement'].str.replace(' cc', '').astype(float)

# Show the cleaned data
cleaned_displacement_data_preview = data.head()
cleaned_displacement_data_preview
```

```python
# Remove 'km' from the Odometer column and convert to integer type
data['Odometer'] = data['Odometer'].str.replace(' km', '').str.replace(',', '').astype(int)

# Show the cleaned data
cleaned_odometer_data_preview = data.head()
cleaned_odometer_data_preview
```

```python
# Replace NaN values in the 'Transmission' column with 'Manual'
data['Transmission'] = data['Transmission'].fillna('Manual')

data
```

```python
# Replace NaN values in the 'Body Type' column with 'Hatchback'
data['Body Type'] = data['Body Type'].fillna('Hatchback')

# Show the updated data
updated_body_type_data_preview = data.head()
updated_body_type_data_preview
```

```python
def adjust_price(price_str):
    try:
        # Remove the 'lacs' and 'crore' text and convert to float
        price_str = price_str.replace('lacs', '').replace('crore', '')
        
        # Split the price range and take the average if it's a range
        prices = price_str.split('-')
        prices = [float(p.strip()) for p in prices if p.strip()]
        avg_price = sum(prices) / len(prices)
        
        # Convert all prices to lacs (1 crore = 100 lacs)
        if 'crore' in price_str:
            avg_price *= 100

        return avg_price
    except:
        return None  # Return None if conversion fails

# Apply the function to the Price column
data['Price'] = data['Price'].apply(adjust_price)

# Show the updated DataFrame
data.head()
```

```python
import pandas as pd
import numpy as np

# Load the data

# Function to clean the displacement values
def clean_displacement(value):
    if isinstance(value, str):
        value = value.replace('cc', '').strip()
        if '-' in value:
            values = value.split('-')
            return np.mean([float(v.strip()) for v in values if v.strip()])
        elif value.isdigit():
            return float(value)
    return np.nan

# Apply the function to clean up the displacement
data['Displacement'] = data['Displacement'].apply(clean_displacement)

# Calculate the mean displacement, ignoring NaNs
mean_displacement = data['Displacement'].mean()

# Replace NaN values with the mean displacement
data['Displacement'].fillna(mean_displacement, inplace=True)

# Save the cleaned data to a new CSV file
data.to_csv('cleaned_data.csv', index=False)

# Display the updated DataFrame
print(data.head())
```

```python
def clean_seating_capacity(value):
    if isinstance(value, str):
        # Remove the word 'Person' and any potential plurals
        value = value.replace('Persons', '').replace('Person', '').strip()
        # Handle ranges by taking the maximum value
        if '-' in value:
            return max([int(v.strip()) for v in value.split('-')])
        return int(value)
    return None  # Return None for unexpected data types

# Apply the function to clean up the seating capacity
data['Seating Capacity'] = data['Seating Capacity'].apply(clean_seating_capacity)

# Save the cleaned data to a new CSV file

# Display the updated DataFrame
print(data.head())
```

```python
def clean_seating_capacity(value):
    if isinstance(value, str):
        # Remove the word 'Persons' or 'Person' and any spaces
        value = value.replace('Persons', '').replace('Person', '').strip()
        # If the value is a range, take the maximum value
        if '-' in value:
            values = [int(v.strip()) for v in value.split('-') if v.strip().isdigit()]
            return max(values) if values else None
        elif value.isdigit():
            return int(value)
    return None  # Handle cases where conversion is not possible

# Apply the function to clean up the seating capacity
data['Seating Capacity'] = data['Seating Capacity'].apply(clean_seating_capacity)

# Save the cleaned data to a new CSV file

# Display the updated DataFrame
print(data.head())
```
